```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend .plotting import plot_decision_regions
```

```python
dataset=pd.read_csv("student_clustering_.csv")
dataset.head(4)
```

- **Loads the dataset**: Reads the student_clustering_.csv file into a pandas DataFrame.  

- **Displays the first 4 rows**: Uses head(4) to show a quick preview of the dataset, helping to understand its structure and contents.

```python
dataset.isnull().sum()
```

- **Checks for missing values in the dataset**: The isnull() function identifies missing (NaN) values in each column.  

- **Counts the number of missing values**: The sum() function aggregates the total number of missing values for each column in the dataset.  

- **Helps in data cleaning**: If any column has missing values, further steps like removing or imputing them can be taken before training a machine learning model.

```python
plt.figure(figsize=(4,2))
sns.kdeplot(dataset["iq"])
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.scatterplot(x="cgpa",y="iq", data=dataset,hue="placed")
plt.show()
```

This code creates a **4x4 scatter plot** using Seaborn to visualize the relationship between **CGPA and IQ**, with different colors representing placement status.

```python
x=dataset.iloc[:,:-1]
y=dataset["placed"]
from sklearn.model_selection import train_test_split
x_test , x_train , y_test , y_train = train_test_split(x,y,test_size = 0.2 , random_state = 42)
```

This code prepares the dataset by selecting input features and the target variable, then splitting it into **80% training** and **20% testing** sets. The random state ensures consistency across runs. However, there is an error in variable assignment, as training and testing variables are swapped, which may lead to confusion later.

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
gnb = GaussianNB()
gnb.fit(x_train,y_train)
gnb.score(x_test,y_test)*100 , gnb.score(x_train,y_train)*100
plt.figure(figsize=(6,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.show()
```

- **Imports Naïve Bayes classifiers**: GaussianNB, MultinomialNB, and BernoulliNB from Scikit-Learn, though only GaussianNB is used.  

- **Creates a Gaussian Naïve Bayes model**: Initializes an instance of GaussianNB(), which is suitable for continuous data.  

- **Trains the model**: Uses the `fit()` function to train the model on the training dataset (x_train, y_train).  

- **Evaluates model accuracy**: Computes and returns the accuracy scores for both the test and training datasets, converting them into percentages.  

- **Visualizes decision boundaries**:  
  - Creates a **6x3 plot** using Matplotlib  
  - Uses plot_decision_regions() to **illustrate the decision boundaries** for the trained model.  
  - Displays the plot to show how the model classifies different regions in the dataset.

```python
mlt = MultinomialNB()
mlt.fit(x_train,y_train)
mlt.score(x_test,y_test)*100 , mlt.score(x_train,y_train)*100
plt.figure(figsize=(6,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mlt)
plt.show()
```

- **Initializes a Multinomial Naïve Bayes model**: Uses MultinomialNB(), which is typically suited for **categorical or count-based** data (e.g., text classification or frequency-based features).  

- **Trains the model**: The fit() function trains the model on the input features (x_train) and target labels (y_train).  

- **Evaluates model accuracy**: Computes the accuracy scores for both the test and training datasets, converting them into percentages.  

- **Visualizes decision boundaries**:  
  - Creates a **6x3 plot** using Matplotlib.  
  - Uses plot_decision_regions() to **display decision boundaries**, showing how the model classifies different regions based on the input data.  
  - Displays the plot to visualize how well the model differentiates between classes.

```python
bnb = BernoulliNB()
bnb.fit(x_train,y_train)
bnb.score(x_test,y_test)*100 , bnb.score(x_train,y_train)*100
plt.figure(figsize=(6,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
plt.show()
```

- **Initializes a Bernoulli Naïve Bayes model**: Uses BernoulliNB(), which is best suited for **binary or boolean feature data** (e.g., text classification with binary word presence).  

- **Trains the model**: The fit() function trains the model using the training dataset (x_train, y_train).  

- **Evaluates model accuracy**: Computes the accuracy scores for both test and training datasets, converting them into percentages to assess performance.  

- **Visualizes decision boundaries**:  
  - Creates a **6x3 plot** using Matplotlib.  
  - Uses plot_decision_regions() to **illustrate classification boundaries**, showing how the model distinguishes between different classes.  
  - Displays the plot to analyze how well the model performs in decision-making.
